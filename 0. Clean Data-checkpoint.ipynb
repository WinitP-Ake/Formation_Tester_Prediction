```python
reduced_test  = test_with_reschart[['WellID', 'Category', 'GR', 'RT', 'DEN', 'NPHI']]
def get_xclass(category):
    return 1 if category=='Normal' else 0    
reduced_test['xclass'] = reduced_test.Category.map(get_xclass)
def get_zone(well_id):
    assert len(well_id) > 3
    return well_id[:3]
reduced_test['zone'] = reduced_test.WellID.map(get_zone)
reduced_test['log_RT'] = np.log(reduced_test.RT)
reduced_test= reduced_test[final_column_order]
reduced_test.dropna()
```

```python
reduced_merge = merged_table[['WellID', 'Category', 'GR', 'RT', 'RHOB', 'NPHI' ]]
reduced_merge['DEN'] = reduced_merge['RHOB']
print(reduced_merge.columns)
reduced_merge['xclass'] = reduced_merge.Category.map(get_xclass)
reduced_merge['zone'] = reduced_merge.WellID.map(get_zone)
reduced_merge['log_RT'] = np.log(reduced_merge.RT)
reduced_merge = reduced_merge[final_column_order]
reduced_merge = reduced_merge.dropna()
```

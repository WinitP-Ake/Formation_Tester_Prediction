```python
# comparison plot
def classifier_check_plot(cl_name: str, cl: 'Classifier', test_fxc: FeatureXClass, sample_name:str):

    prob = cl.predict_proba(test_fxc.features)[:, 1]
    pred_fxc = FeatureXClass(features=test_fxc.features, xclass=prob)

    all_test_fxc = test_fxc

    fig, axes = plt.subplots(figsize=(40,40), ncols=2, nrows=2)
    top_features = ('GR', 'DEN')
    bottom_features = ('log_RT', 'NPHI')

    def make_plot(features, fxc):
        x_features, y_features = features
        prob_plot(fxc, x_features, y_features, with_text=True)

    def center_text(text, fontsize=24):
        plt.text(0.5, 0.5, text,
                 horizontalalignment='center',
                 verticalalignment='center',
                 fontsize=fontsize,
                 transform = plt.gca().transAxes, alpha=0.4)

    plt.sca(axes[0,0])
    make_plot(top_features, pred_fxc)
    center_text(f'{cl_name}\n{top_features}', fontsize=60)


    plt.sca(axes[0,1])
    make_plot(top_features, all_test_fxc)
    center_text(sample_name, fontsize=60)

    plt.sca(axes[1,0])
    make_plot(bottom_features, pred_fxc)
    center_text(f'{cl_name}\n{bottom_features}', fontsize=60)

    plt.sca(axes[1,1])
    make_plot(bottom_features, all_test_fxc)
    center_text(sample_name, fontsize=60)

    plt.tight_layout()
```

```python
classifier_check_plot('sensile_lgbm', classifiers['sensible_lgbm'], all_sample.test_sample, sample_name='Test Only')
```

```python
for cl_name, cl in classifiers.items():
    print(cl_name)
    classifier_check_plot(cl_name, cl, all_sample.test_sample, sample_name='Test Only')
    plt.savefig(f'cl_check_{cl_name}.pdf', bbox_inches='tight')
    plt.savefig(f'cl_check_{cl_name}.png', bbox_inches='tight')
    plt.close()
```

```python
for cl_name, cl in classifiers.items():
    print(cl_name)
    classifier_check_plot(cl_name, cl, all_data_fxc, sample_name='All Data(Train+Test)')
    plt.savefig(f'cl_check_full_{cl_name}.pdf', bbox_inches='tight')
    plt.savefig(f'cl_check_full_{cl_name}.png', bbox_inches='tight')
    plt.close()
```

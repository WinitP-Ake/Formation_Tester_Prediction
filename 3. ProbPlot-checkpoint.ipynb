```python
import scipy.ndimage
from matplotlib import pyplot as plt
from wtn_mdt.data_loader import load_all_data_df
from wtn_mdt.prob_plot import prob_plot
from wtn_mdt.const import DEFAULT_FEATURES
from wtn_mdt.feature_xclass import FeatureXClass
from matplotlib import pyplot as plt
```

```python
import numpy as np
import math
```

## Load data

```python
# load all data
all_data_df = load_all_data_df()
print(all_data_df.shape)
all_data_df.head()
```

```python
# remove lost_seal from category
all_data_df = all_data_df[all_data_df['Category'].isin(['Normal','Dry','Tight','Unstable','Limited_drawdown','Superchange'])]
all_data_df.shape
```

```python
all_data_df['Category'].value_counts()
```

```python
# load feature & class as objects
all_feature_xclass = FeatureXClass.from_df(all_data_df, features=DEFAULT_FEATURES)
print('-- Features --')
print(all_feature_xclass.features.head())
print('-- Class --')
print(all_feature_xclass.xclass.head())
```

## Visualization of the result

#### 1. Plot of GR and DEN

```python
# using 
# from wtn_mdt.prob_plot import prob_plot

plt.figure(figsize=(20,20))
prob_plot(data = all_feature_xclass,
         x_features = 'GR',
         y_features = 'DEN', #'DEN'
         cut_off=0.30
         )

#plt.show()
plt.savefig('prob_plot_gr_den.pdf', bbox_inches='tight')
```

#### 2. Plot of NPHI and log_RT

```python
plt.figure(figsize=(20,20))
prob_plot(all_feature_xclass, 'NPHI', 'log_RT', cut_off=0.30, )
plt.savefig('prob_plot_NPHI_logRT.pdf', bbox_inches='tight')
```

## Looping plot

```python
from itertools import combinations
for f1, f2 in combinations(DEFAULT_FEATURES, 2):
    print(f1, f2)
    plt.figure(figsize=(20,20))
    prob_plot(all_feature_xclass, f1, f2, cut_off=0.30, )
    plt.savefig(f'prob_plot_{f1}_{f2}.pdf', bbox_inches='tight')
    plt.savefig(f'prob_plot_{f1}_{f2}.png', bbox_inches='tight')
    plt.close()
```

```python
areas_sample = get_train_test_sample_dict(all_data_df)

areas_classifiers = create_and_train_classifier_for_all_areas(areas_sample)
```

```python
# get perf summary for pair of train test
area_perf_summary = {}
for train_area in areas_sample.keys():
    for test_area in areas_sample.keys():
        classifiers = areas_classifiers[train_area]
        sample = areas_sample[test_area]
        perf_summary = perf.get_perf_summary(classifiers, sample.test_sample)
        area_perf_summary[(train_area, test_area)]=perf_summary
```

```python
n_train_areas = len(areas_sample)
n_test_areas = len(areas_sample)
fig, ax = plt.subplots(figsize=(20,20), nrows=n_train_areas, ncols=n_test_areas)
for i_train, train_area in enumerate(areas_sample.keys()):
    for i_test, test_area in enumerate(areas_sample.keys()):
        perf_summary = area_perf_summary[train_area, test_area]
        benchmark = areas_sample[test_area].get_test_benchmark()
        plt.axes(ax[i_train, i_test])
        plot_bar_comparison(perf_summary, benchmark)
        best_acc = perf_summary['logit_DEN'].best_acc
        plt.text(0.5, 0.5, f'train {train_area}/test {test_area}\n {best_acc:.2f} / {benchmark:.2f}',
             horizontalalignment='center',
             verticalalignment='center',
             fontsize=24,
             transform = plt.gca().transAxes, alpha=0.4)
plt.subplots_adjust(wspace=0.4)        
plt.savefig('cross_area_test.pdf', bbox_inches='tight')
```
